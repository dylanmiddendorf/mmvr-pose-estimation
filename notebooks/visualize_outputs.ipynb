```python
import matplotlib.pyplot as plt
import numpy as np

connections = np.array([[13, 15], [11, 13], [14, 16], [12, 14], [11, 12],
                [5, 11], [6, 12], [5, 6], [5, 7], [6, 8], [7, 9],
                [8, 10], [1, 2], [0, 1], [0, 2], [1, 3], [2, 4],
                [3, 5], [4, 6]])

def compare_output(pred, gt, guess):
    fig = plt.figure(figsize=(15, 12))
    ax = fig.add_subplot(1,1,1)
    img = np.zeros((480, 640, 3), dtype=np.uint8)

    for connection in connections:
        x_gt = gt[guess,connection, 0]
        x_pred = pred[guess, connection, 0]
        y_gt = gt[guess,connection, 1]
        y_pred = pred[guess, connection, 1]
        ax.plot(x_pred, y_pred,color='r',marker='.')
        ax.plot(x_gt, y_gt,color='b',marker='.')
    ax.imshow(img)
    plt.show()

sample_output = np.load("../sample-output.npz")
guess, truth = sample_output["guess"], sample_output["truth"]
for i in range(4):
    compare_output(guess, truth, i)
```
